**Unsupervised Machine Learning Clustering Algorithm: K-Means (Base Model) | Metrics Used - Euclidean Distance**

**Objective:**
The objective of this analysis is to segment consumers based on their similar characteristics using the K-Means clustering algorithm.

**Approach:**
*K-Means Clustering:* K-Means is an unsupervised machine learning algorithm that partitions data points into clusters based on similarity. It aims to minimize the sum of squared distances (inertia) within each cluster. The algorithm iteratively assigns data points to the nearest cluster centroid and updates the centroids until convergence.

**Metrics Used - Euclidean Distance:**
Euclidean distance measures the straight-line distance between two data points in the feature space. It is commonly used as the distance metric in K-Means. The algorithm minimizes the sum of squared Euclidean distances within each cluster.

**Interpretation:**
The K-Means algorithm will group similar consumers together based on their features (e.g., original price, ratings, reviews, discount price, etc.). The choice of K (number of clusters) will impact the segmentation results. The Euclidean distance metric ensures that data points within the same cluster are close to each other.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Assuming df_cluster_0, df_cluster_1, and df_cluster_2 are DataFrames containing the ratings for each cluster

# Combine all ratings into a single list with cluster labels
all_ratings = pd.concat([df_cluster_0['Ratings'], df_cluster_1['Ratings'], df_cluster_2['Ratings']])
labels = ['Cluster 0'] * len(df_cluster_0) + ['Cluster 1'] * len(df_cluster_1) + ['Cluster 2'] * len(df_cluster_2)

# Perform Tukey's HSD test
tukey_results = pairwise_tukeyhsd(all_ratings, labels)

# Print the pairwise comparison results
print(tukey_results)

# Convert the results to a DataFrame for visualization
tukey_df = pd.DataFrame(data=tukey_results._results_table.data[1:], columns=tukey_results._results_table.data[0])

# Pivot the DataFrame for heatmap
tukey_pivot = tukey_df.pivot(index='group1', columns='group2', values='meandiff')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(tukey_pivot, annot=True, cmap="YlGnBu", fmt=".3f")
plt.title("Tukey's HSD Mean Difference Comparison")
plt.xlabel("Cluster")
plt.ylabel("Cluster")
plt.show()
```
